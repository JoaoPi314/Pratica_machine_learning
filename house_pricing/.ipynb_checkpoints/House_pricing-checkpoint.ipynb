# 1 - Algoritmo para prever preço de casas

## 1.1 Descrição do dataset

**O dataset utilizado será o dataset disponibilizado pelo scikit learn: Boston dataset.**
**O dataset contém 13 features e 506 colunas. Segue a descrição de cada feature:**



1. CRIM: Índice criminal urbano por cabeça
2. ZN: Proporção de terreno residencial que excede 25000 pés quadrados
3. INDUS: Proporção de terrenos comerciais não varejistas em uma cidade
4. CHAS: Charles River (1 indica que os limites são um rio, 0 caso contrário)
5. NOX: Concentração de óxido nítrico
6. RM: Número médio de salas em uma casa
7. AGE: Proporção de casas particulares completadas antes de 1940
8. DIS: Distância com pesos para as 5 regiões centrais de Boston
9. RAD: Índice de proximidade de uma rodovia radial
10. PTRATIO: Proporção de professores e estudantes em uma área urbana
11. target: Preço médio de uma casa particular: 1 equivale à $1000

**Vamos começar importando as bibliotecas e funções necessárias**

```python
#Para prevenir warnings nas saídas
import warnings
warnings.filterwarnings('ignore')
```

```python
#Conjunto de pacotes básicos da ciência de dados
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns
```

```python
#Define atributos para prevenir caracteres chineses
mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus']=False
```

```python
#Introdução dos pacotes de machine learning, pré-processamento, seleção de modelo e métricas
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
```

```python
#Importando o Boston dataset
from sklearn.datasets import load_boston
```

```python
#Algoritmos a serem utilizados
from sklearn.linear_model import RidgeCV, LassoCV, LinearRegression, ElasticNet
#Essa é a forma de regressão de SVM
from sklearn.svm import SVR
#Algoritmos ensemble
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

## 1.2 Visuaização dos dados

```python
boston = load_boston()

x = boston.data
y = boston.target
```

```python
print('Nomes das features: \n{}'.format(boston.feature_names))
print('Tamanho da amostra de dados: {}'.format(x.shape))
print('Tamanho da amostra alvo: {}'.format(y.shape))
```

```python
#Para converter para dataframe
x = pd.DataFrame(boston.data, columns=boston.feature_names)
x.head()
```

```python
x.isnull().sum()
```

**Abaixo, podemos observar a distribuição dos preços das casas**

```python
sns.distplot(tuple(y), kde=False, fit=st.norm)
```

## 1.3 Pré-processamento dos dados

```python
#Separando os dados com train_test_split (20% dos dados destinados a teste)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=28)
```

```python
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
x_train
```

## 1.4 Aplicação de vários algoritmos de regressão

```python

nomes= ['LinearRegression',
        'Ridge',
        'Lasso',
        'Random Forest',
        'GBDT',
        'Support Vector Regressor',
        'ElasticNet',
        'Xgboost']
#Definição do modelo (CV = Cross validation)

models = [LinearRegression(),
          RidgeCV(alphas=(0.001,0.1,1), cv=3),
          LassoCV(alphas=(0.001,0.1,1), cv=5),
          RandomForestRegressor(n_estimators=10),
          GradientBoostingRegressor(n_estimators=30),
          SVR(),
          ElasticNet(alpha=0.001, max_iter=10000),
          XGBRegressor()]
```

**Função que irá retorna o score R² de cada modelo de regressão aplicado**

```python
def R2(model, x_train, x_test,y_train, y_test):
    model_fitted = model.fit(x_train,y_train)
    y_pred = model_fitted.predict(x_test)
    score = r2_score(y_test, y_pred)
    return score
```

```python
for nome,model in zip(nomes, models):
    score = R2(model, x_train, x_test, y_train, y_test)
    print('{}:{:.6f}, {:.4f}'.format(nome, score.mean(), score.std()))
```

## 1.5 Hiperparâmetros e Grid Search

* Kernel: Função Kernel
* C: Fator de regularização SVR
* Gamma: 'rbf', 'poly', 'sigmoid' coeficientes da função kernel, que afetam a performance do modelo

```python
parameters = {
    'kernel':['linear', 'rbf'],
    'C':[0.1, 0.5, 0.9, 1, 5],
    'gamma':[0.001, 0.01, 0.1, 1]
}

#Grid search e cross validation

model = GridSearchCV(SVR(), param_grid = parameters, cv=3)
model.fit(x_train, y_train)
```

```python
print('Melhores parâmetros: {}'.format(model.best_params_))
print('Melhor modelo:{}'.format(model.best_estimator_))
print('Melhor score R²: {}'.format(model.best_score_))
```

## 1.6 Visualização da performance do modelo

```python
ln_x_test = range(len(x_test))
y_predict = model.predict(x_test)
```

```python
plt.figure(figsize=(16,8), facecolor='w')
plt.plot(ln_x_test, y_test, 'r-', lw=2, label=u'Value')
plt.plot(ln_x_test, y_predict, 'g-', lw=3, label=u'Valor estimado do SVR. $R^2$=%.3f' %(model.best_score_))

plt.legend(loc='upper left')
plt.grid(True)
plt.title(u'Previsão de preços de casas de Boston (SVR)')
plt.xlim(0, 101)
plt.show()
```
